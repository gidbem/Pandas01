```python
import numpy as np

#criando um array apartir de uma lista

#Para criar uma rray, todos os elementos tem que ser do mesmo tipo de dado
lista = [1,2,3,4,5]
array_np = np.array(lista)

print(f'Lita python: {lista}')
print(f'Array numpy: {array_np}')

#Somar 5 a todos os elementos do array 
array_np += 5


print(f'Add (+5) ==> {array_np}')

#Criando matriz
#Todas as dimensões da matriz devem possuir o mesmo tamanho
matriz = np.array([[1, 2, 3, 4, 5], [10, 11, 12, 13, 14]])

print(matriz)
```

```python
#Pandas
import pandas as pd

#Series
#Lista de notas python
notas = [8.5, 7.3, 6, 8, 10]

#Transformando a lista em uma Serie
serie_notas = pd.Series(notas)
print(serie_notas)
```

```python
series_notas_alunos = pd.Series([8.5, 7.3, 6, 8, 10, 9],
                                index=['Juca', 'Ana', 'Pedro', 'Silvio', 'Giovanna', 'Fernanda'])

print(series_notas_alunos)
```

```python
#Exibindo a nota de indice Giovanna
nome = input('Informe um nome:')

print(f'{nome}: {series_notas_alunos[nome]}')
```

```python
#DataFrame
#Para criar um DataFrame normalmente trabahamos com a estrutura de dicionário
#Onde as chaves são as colunas e os valores são dados
dados_alunos = {
    'Nome' : ['Ana', 'Mariano', 'Juca', 'Fernanda', 'Gabi'],
    'Idade' : [20, 23, 21, 32, 30],
    'Nota' : [8.8, 6.4, 7.2, 9.5, 10],
    'Aprovado' : [True, False, True, True, True]
}

df_dados_alunos = pd.DataFrame(dados_alunos)
print(df_dados_alunos)
```
